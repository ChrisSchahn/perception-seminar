## 1. Einleitung

Im Social Media Raum, vor allem Instagram, wird viel mit Filtern und Bildbearbeitung gearbeitet, um eine vermeintlich bessere Bildqualität beim Konsumenten zu suggerieren. Unsere Forschungsfrage war, ob sich Social Media Konsum auf die Wahrnehmung von Bildbearbeitungsfiltern auswirkt. Dabei haben wir die folgenden zwei Hypothesen getestet: 

**1) Fotos mit Filtern werden von Langzeit Social Media Nutzern als optisch ansprechender empfunden**

**2) Filter in Fotos werden von Langzeit Social Media Nutzern eher erkannt als von nicht Social Media Nutzern**

Um unsere Hypothesen und Fragestellung zu testen, haben wir zwei Experimente durchgeführt, bei denen wir Bilder mit Instagram Filtern bearbeitet haben. Bei dem ersten Experiment haben wir in verschiedenen Variationen zweimal dasselbe Motiv verglichen, das jeweils unterschiedlich stark bearbeitet wurde, wobei der Proband das für ihn optisch ansprechendere auswählen sollte. Beim zweiten Experiment war mehrmals hintereinander ein Bild zu sehen, das unterschiedlich stark bearbeitet war und der Proband sollte auf einer Skala einschätzen, wie stark das Bild bearbeitet war. Um die Experimente durchzuführen, haben wir eigenen Python Code entwickelt sowie den von Guillermo Aguilar zur Verfügung gestellten Python Code verwendet und angepasst für unseren Versuch.

Wir haben außerdem vier frei zugängliche Motive verwendet und diese mit drei verschiedenen Instagram Filtern bearbeitet, **Clarendon**, **Juno** und **Lark**. Jedes Motiv wurde mit allen Filtern und den Intensitätsstufen 0%, 25%, 50%, 75% und 100% bearbeitet, die Bearbeitung fand mithilfe der Instagram App statt.

```python

def add_name_and_response(df, name):
    image_name = df['image_a'].str.split("_", n = 1, expand = True)[0]
    df.insert(1, 'name', image_name)
    df['observer'] = name
    return df


def compute_means_and_std(d_frame, observers):
    user = d_frame
    user_count = {'0': [], '25': [], '50': [], '75': [], '100': []}

    for observer in observers:
        u_name = user.loc[user['observer'] == observer]
        (unique, user_counts) = np.unique(u_name['chosen_i'], return_counts=True)

        for intens, count in zip(unique, user_counts):
            user_count[str(intens)].append(count)

    user_means, user_std = [], []

    keys = ['0', '25', '50', '75', '100']
    
    for key in keys:
        user_means.append(np.mean(user_count[key]))
        user_std.append(np.std(user_count[key]))


    return np.asarray(user_means), np.asarray(user_std)
```

```python
def add_name_and_response(df, name):
    image_name = df['image'].str.split("_", n = 1, expand = True)[0]
    df['observer'] = name
    df.insert(4, 'response_i', df['response'])
    df.insert(2, 'name', image_name)
    df.loc[df['response'] == 1, 'response_i'] = 0
    df.loc[df['response'] == 2, 'response_i'] = 25
    df.loc[df['response'] == 3, 'response_i'] = 50
    df.loc[df['response'] == 4, 'response_i'] = 75
    df.loc[df['response'] == 5, 'response_i'] = 100
    return df
```

```python
def compute_means_and_std(d_frame, observers):
    user = d_frame
    user_count = {'0': [], '25': [], '50': [], '75': [], '100': []}

    for intens in range(0,101,25):
        u_name = user.loc[user['intensity'] == intens]
        (unique, user_counts) = np.unique(u_name['response_i'], return_counts=True)
        
        #for intens, count in zip(unique, user_counts):
        user_count[str(intens)].append(u_name['response_i'].tolist())
        
            

    user_means, user_std = [], []

    keys = ['0', '25', '50', '75', '100']
    
    for key in keys:
        user_means.append(np.mean(user_count[key]))
        user_std.append(np.std(user_count[key]))
        #user_std.append(np.var(user_count[key]))


    return np.asarray(user_means), np.asarray(user_std)
```

## 4. Diskussion

Wir können sagen, dass unsere Fragestellung hinreichend beantwortet wurde und wir eine Auswirkung von Social Media Konsum auf die Wahrnehmung festgestellt haben. Jedoch sprechen wir bei der Auswertung unserer Hypothesen nur von Trends in die eine oder andere Richtung, da wir eine sehr geringe Versuchsteilnehmer hatten und so kein klares Ergebnis erhalten. Dazu ist eine Umfrage in größerem Stil notwendig.

Die erste Hypothese,

**1) Fotos mit Filtern werden von Langzeit Social Media Nutzern als optisch ansprechender empfunden**,

sehen wir als "belegt" an, nach Auswertung unserer Ergebnisse vom ersten Versuch. **//Chris kurze Erklärung einfügen**

Die zweite Hypothese,

**2) Filter in Fotos werden von Langzeit Social Media Nutzern eher erkannt als von nicht Social Media Nutzern**,

sehen wir als "widerlegt" an. Das zweite Experiment hat deutlich gemacht, dass hoher Social Media Konsum keine bessere Wahrnehmung von Filtern voraussetzt. Diese Einschätzung wird durch die obigen Graphiken nochmal verdeutlicht. 

Überraschend war, dass trotz der geringen Unterscheidung beider Gruppen zwischen 0% und 25% Intensität, eindeutig Bilder mit 25% gegenüber 0% bevorzugt wurde. Begründung?


### Probleme

Durch die geringe Anzahl Teilnehmeranzahl und die niedrige Anzahl verwendeter Stimuli ist keine klare Aussage bezüglich unserer Fragestellung möglich. Außerdem war unsere frei zugängliche Datenbank eingeschränkt im geeigneten Bildmaterial. Die Erhöhung der Stimuli würde zu einem genaueren Ergebnis führen, jedoch auch die bereits von den Teilnehmern als lang empfundene Dauer des Experiments weiter stark erhöhen. Das Experiment wurde noch dazu als _eintönig_ beschrieben. Wir haben das Experiment lokal auf unseren Computern durchgeführt, was die Anzahl möglicher Teilnehmer stark eingeschränkt hat.

### Verbesserungsvorschläge

Offensichtlich würde eine höhere Anzahl Teilnehmer sowie Stimuli das Ergebnis deutlich verbessern und eine genauere Aussage ermöglichen. Dazu müsste das Experiment auch besser verfügbar sein, beispielsweise als online Umfrage.

### Ausblick

Insgesamt können wir feststellen, dass Social Media Konsum einen nicht zu vernachlässigenden Faktor in der Wahrnehmung von Bildern darstellt. Bei einer Anzahl von 4,2 Milliarden aktiven Social Media Nutzern im Jahre 2021 $^1$ sollte dies in Studien zu Wahrnehmung und Bildqualität unbedingt berücksichtigt werden.
